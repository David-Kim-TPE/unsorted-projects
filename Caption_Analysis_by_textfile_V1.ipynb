# Run Filename to Caption

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def generate_wordcloud(file_path):
    """
    Generates a word cloud from a text file located at the given file path.
    Each line in the file should be an image caption.

    Parameters:
    - file_path (str): The path to the text file containing image captions.

    Returns:
    - An image of the word cloud visualization, with words sized according to their frequency in the text.
    """

    # Read the entire content of the file
    with open(file_path, 'r') as file:
        text = file.read()

    # Create a WordCloud object
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    # Display the word cloud using matplotlib
    plt.figure(figsize=(8, 4))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Example usage
# To use this function, uncomment the line below and replace 'path/to/text_file.txt' with the actual file path
generate_wordcloud(root_folder)
```
