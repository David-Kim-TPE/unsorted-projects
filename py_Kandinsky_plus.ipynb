```python
from PIL import Image, ImageDraw
import random

def generate_kandinsky_style_art(width, height):
    # Create a blank canvas
    image = Image.new("RGB", (width, height), "white")
    draw = ImageDraw.Draw(image)

    # Define a set of colors reminiscent of Kandinsky's palette
    colors = ['#69D2E7', '#A7DBD8', '#E0E4CC', '#F38630', '#FA6900', '#FF4E50', '#F9D423']

    # Generate random shapes
    for _ in range(100):  # Adjust the number of shapes as needed
        shape_type = random.choice(['circle', 'rectangle', 'line'])
        color = random.choice(colors)

        if shape_type == 'circle':
            radius = random.randint(10, 100)  # Random radius
            x = random.randint(0, width)
            y = random.randint(0, height)
            draw.ellipse((x-radius, y-radius, x+radius, y+radius), fill=color, outline=None)

        elif shape_type == 'rectangle':
            x1 = random.randint(0, width)
            y1 = random.randint(0, height)
            x2 = random.randint(x1, width)
            y2 = random.randint(y1, height)
            draw.rectangle((x1, y1, x2, y2), fill=color, outline=None)

        elif shape_type == 'line':
            x1 = random.randint(0, width)
            y1 = random.randint(0, height)
            x2 = random.randint(0, width)
            y2 = random.randint(0, height)
            draw.line((x1, y1, x2, y2), fill=color, width=5)  # Random lines

    return image

# Generate an art piece in the style of Wassily Kandinsky
kandinsky_style_art = generate_kandinsky_style_art(1024, 1024)
# Display the art
kandinsky_style_art.save("kandinsky_style_001.png")
```

```python
import numpy as np
from PIL import Image, ImageDraw
from datetime import datetime

# Function to generate the Mandelbrot set
def mandelbrot_zoomed_edge(width, height, x_min, x_max, y_min, y_max, max_iter):
    x = np.linspace(x_min, x_max, num=width).reshape((1, width))
    y = np.linspace(y_min, y_max, num=height).reshape((height, 1))
    C = np.tile(x, (height, 1)) + 1j * np.tile(y, (1, width))

    Z = np.zeros(C.shape, dtype=complex)
    M = np.full(C.shape, True, dtype=bool)
    for i in range(max_iter):
        Z[M] = Z[M] * Z[M] + C[M]
        M[np.abs(Z) > 2] = False
    return M

# Function to apply an inverted gradient from dark grey to white
def apply_inverted_gradient(fractal, width, height):
    image = Image.new("RGB", (width, height))
    pixels = image.load()

    color_top = (64, 64, 64)  # Dark grey
    color_bottom = (255, 255, 255)  # White
    gradient_steps = [tuple(int(color_top[j] + (color_bottom[j] - color_top[j]) * (i / height)) for j in range(3)) for i in range(height)]

    for y in range(height):
        inverted_gradient_color = gradient_steps[y]
        for x in range(width // 2):
            if fractal[y, x]:
                pixels[x, y] = inverted_gradient_color
                pixels[width - x - 1, y] = inverted_gradient_color
    return image

# Function to generate a unique filename based on the current timestamp
def generate_unique_filename_png():
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    return f"fractal_art_{timestamp}.png"

# Parameters for the zoomed-in fractal
width, height = 1024, 1024
zoom_x_min, zoom_x_max = -0.74877, -0.74872
zoom_y_min, zoom_y_max = 0.065053, 0.065103
zoom_max_iter = 500

# Generate the fractal
fractal_zoomed_edge = mandelbrot_zoomed_edge(width // 2, height, zoom_x_min, zoom_x_max, zoom_y_min, zoom_y_max, zoom_max_iter)

# Apply the inverted gradient to the fractal
image_with_inverted_gradient = apply_inverted_gradient(fractal_zoomed_edge, width, height)

# Save the image with a unique filename
unique_filename = generate_unique_filename_png()
image_with_inverted_gradient.save(f"/content/{unique_filename}", "PNG")

# This would be the output filename
print(unique_filename)
```

# Final Branching Generator

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

# Function to generate a unique filename based on the current timestamp
def generate_unique_filename_png():
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    return f"algorithmic_botanical_{timestamp}.png"

# Define the L-system rules
rules = {
    "X": "F+[[X]-X]-F[-FX]+X",
    "F": "FF"
}

# Function to apply rules to generate the L-system string
def apply_rules(axiom, iterations):
    for _ in range(iterations):
        new_axiom = "".join(rules.get(char, char) for char in axiom)
        axiom = new_axiom
    return axiom

# Function to interpret L-system string and return line segment data
def interpret_l_system(axiom, angle, length):
    directions = np.array([0, 1])  # Initial direction is "up"
    current_position = np.array([0.0, 0.0])  # Starting at the origin
    stack = []
    lines = []  # List to hold line segment start and end points

    for cmd in axiom:
        if cmd == 'F':
            start_position = current_position.copy()
            current_position += directions * length
            lines.append((start_position, current_position.copy()))
        elif cmd == '+':
            directions = np.dot(directions, [[np.cos(np.radians(angle)), -np.sin(np.radians(angle))],
                                            [np.sin(np.radians(angle)), np.cos(np.radians(angle))]])
        elif cmd == '-':
            directions = np.dot(directions, [[np.cos(np.radians(-angle)), -np.sin(np.radians(-angle))],
                                            [np.sin(np.radians(-angle)), np.cos(np.radians(-angle))]])
        elif cmd == '[':
            stack.append((current_position.copy(), directions.copy()))
        elif cmd == ']':
            current_position, directions = stack.pop()

    return lines

# Function to create the plot with the specified size and color, and save it to a file
def create_and_save_plot(lines, image_size, line_color, file_path):
    fig, ax = plt.subplots(figsize=(image_size[0]/100, image_size[1]/100))
    for line in lines:
        start, end = line
        ax.plot([start[0], end[0]], [start[1], end[1]], color=line_color)
    ax.set_aspect('equal')
    plt.axis('off')
    plt.savefig(file_path, bbox_inches='tight', pad_inches=0)
    plt.close(fig)  # Close the figure to free memory

# Start with an axiom
axiom = "X"

# Number of iterations
iterations = 5

# The angle for turns
angle = 25

# The length of a forward move
length = 5

# Define the size of the image (width, height) in pixels and line color
image_size = (1024, 1024)  # Size in pixels
line_color = "#00ff00"  # Green color in hex

# Apply L-system rules to generate the string
l_system_string = apply_rules(axiom, iterations)

# Interpret the L-system string into lines
lines = interpret_l_system(l_system_string, angle, length)

# Generate a unique filename
unique_filename = generate_unique_filename_png()

# Full path for saving the image
file_path = f"/content/{unique_filename}"

# Create the plot and save it to the file
create_and_save_plot(lines, image_size, line_color, file_path)

print(f"Saved algorithmic botanical image as {unique_filename}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

def generate_unique_filename_png(prefix="Mandala"):
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    return f"{prefix}_{timestamp}.png"

def draw_shape(ax, center, radius, num_sides, rotation_angle, color, line_thickness):
    """
    Draws a custom shape centered at a given point. Shapes can be circles, polygons, etc.
    """
    if num_sides < 3:
        shape = plt.Circle(center, radius, color=color, fill=False, linewidth=line_thickness)
    else:
        theta = np.linspace(0, 2*np.pi, num_sides+1) + np.radians(rotation_angle)
        x = center[0] + radius * np.cos(theta)
        y = center[1] + radius * np.sin(theta)
        shape = plt.Polygon(np.vstack((x, y)).T, closed=True, color=color, fill=False, linewidth=line_thickness)
    ax.add_artist(shape)

def draw_mandala(num_layers, num_elements_per_layer, colors, image_size, line_thickness, background_color):
    fig, ax = plt.subplots(figsize=(image_size / 100, image_size / 100))
    ax.set_facecolor(background_color)
    ax.axis('equal')
    plt.axis('off')

    max_radius = image_size / 2 * 0.9
    radii = np.linspace(10, max_radius, num_layers)
    rotation_angle = 0

    for i, radius in enumerate(radii, start=1):
        num_sides = num_elements_per_layer * i  # Increase the number of sides for each layer
        for j in range(num_elements_per_layer):
            theta = 2 * np.pi / num_elements_per_layer * j
            center = (image_size / 200 + radius / 100 * np.cos(theta),
                      image_size / 200 + radius / 100 * np.sin(theta))
            draw_shape(ax, center, radius / 200, num_sides, rotation_angle, np.random.choice(colors), line_thickness)
            rotation_angle += 5  # Increment the rotation angle for a dynamic effect

    ax.set_xlim(0, image_size / 100)
    ax.set_ylim(0, image_size / 100)

    filename = generate_unique_filename_png()
    file_path = f"/content/{filename}"
    plt.savefig(file_path, bbox_inches='tight', pad_inches=0, dpi=100)
    plt.close(fig)

    return filename

# Parameters
num_layers = 6
num_elements_per_layer = 12
colors = ['#ff6347', '#4682b4', '#32cd32', '#ff1493', '#ffa500']
image_size = 1024
line_thickness = 2
background_color = "#ffffff"

mandala_filename = draw_mandala(num_layers, num_elements_per_layer, colors, image_size, line_thickness, background_color)

print(f"Saved mandala image as {mandala_filename}")
```

# Final repeating patterns

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

def generate_unique_filename_png(prefix="RandomizedPattern"):
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    return f"{prefix}_{timestamp}.png"

def create_base_tile(tile_size, color, line_thickness, num_lines):
    """
    Create a single tile with randomized lines that can tile seamlessly.
    Mirrors any lines crossing the edges to ensure seamless tiling.
    """
    lines = []
    for _ in range(num_lines):
        # Start point
        x1, y1 = np.random.uniform(0, tile_size, 2)
        # End point
        x2, y2 = np.random.uniform(0, tile_size, 2)
        lines.append(((x1, y1), (x2, y2)))

        # If the line crosses a boundary, create a mirrored line on the opposite edge
        if x1 == x2 or y1 == y2:  # Simple case: horizontal or vertical lines
            continue  # No need to mirror perfectly horizontal or vertical lines
        # Mirror horizontally
        if x1 < tile_size / 2:
            lines.append(((x1 + tile_size, y1), (x2 + tile_size, y2)))
        else:
            lines.append(((x1 - tile_size, y1), (x2 - tile_size, y2)))
        # Mirror vertically
        if y1 < tile_size / 2:
            lines.append(((x1, y1 + tile_size), (x2, y2 + tile_size)))
        else:
            lines.append(((x1, y1 - tile_size), (x2, y2 - tile_size)))

    return lines

def draw_pattern(tile_size, num_tiles_x, num_tiles_y, color, line_thickness, num_lines):
    fig, ax = plt.subplots(figsize=(num_tiles_x, num_tiles_y))
    ax.set_aspect('equal')
    plt.axis('off')

    # Generate the base tile
    base_tile_lines = create_base_tile(tile_size, color, line_thickness, num_lines)

    # Repeat the base tile across the specified area
    for i in range(num_tiles_x):
        for j in range(num_tiles_y):
            for line in base_tile_lines:
                start, end = line
                plt.plot([start[0] + i * tile_size, end[0] + i * tile_size],
                         [start[1] + j * tile_size, end[1] + j * tile_size], color=color, linewidth=line_thickness)

    filename = generate_unique_filename_png()
    file_path = f"/content/{filename}"
    plt.savefig(file_path, bbox_inches='tight', pad_inches=0, dpi=100)
    plt.close(fig)

    return filename

# Parameters
tile_size = 100
num_tiles_x = 5
num_tiles_y = 5
color = "#6495ED"  # Example color
line_thickness = 2
num_lines = 10  # Number of lines in each tile

# Generate and save the pattern
pattern_filename = draw_pattern(tile_size, num_tiles_x, num_tiles_y, color, line_thickness, num_lines)

print(f"Saved pattern image as {pattern_filename}")
```

# Final Parametric 2 OBJ

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from datetime import datetime

def generate_unique_filename(base="model", ext=""):
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    return f"{base}_{timestamp}.{ext}"

def generate_parametric_model_with_preview(resolution=30, scale=1.0, amplitude=0.5, frequency=1.0, phase=0.0):
    # Generate unique filenames
    obj_filename = generate_unique_filename(base="parametric_model", ext="obj")
    png_filename = generate_unique_filename(base="parametric_model", ext="png")

    u = np.linspace(0, 2 * np.pi, resolution)
    v = np.linspace(0, np.pi, resolution)
    u, v = np.meshgrid(u, v)

    # Parametric equations incorporating controls
    x = scale * (np.sin(v) * np.cos(u))
    y = scale * (np.sin(v) * np.sin(u))
    z = scale * (np.cos(v) + amplitude * np.sin(frequency * u + phase))

    vertices = np.vstack([x.flatten(), y.flatten(), z.flatten()]).T
    faces = generate_faces(resolution, resolution)

    # Save the OBJ file
    save_to_obj(f"/content/{obj_filename}", vertices, faces)

    # Create and save the preview image
    create_preview_image(f"/content/{png_filename}", x, y, z)

    return obj_filename, png_filename

def generate_faces(width, height):
    faces = []
    for i in range(height - 1):
        for j in range(width - 1):
            v1 = i + j * width + 1
            v2 = i + (j + 1) * width + 1
            v3 = i + j * width + 2
            v4 = i + (j + 1) * width + 2
            faces.append([v1, v2, v3])
            faces.append([v2, v4, v3])
    return faces

def save_to_obj(filepath, vertices, faces):
    with open(filepath, "w") as file:
        for vertex in vertices:
            file.write(f"v {vertex[0]} {vertex[1]} {vertex[2]}\n")
        for face in faces:
            file.write(f"f {face[0]} {face[1]} {face[2]}\n")

def create_preview_image(filepath, x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, color='orange', edgecolor='none')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    plt.savefig(filepath)
    plt.close(fig)

# Example usage
obj_filename, png_filename = generate_parametric_model_with_preview(
    resolution=50,
    scale=1,
    amplitude=0.5,
    frequency=2,
    phase=np.pi/4
)

print(f"Model saved as /content/{obj_filename}")
print(f"Preview image saved as /content/{png_filename}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

def generate_unique_filename_png(prefix="RandomizedPattern"):
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    return f"{prefix}_{timestamp}.png"

def create_base_tile(tile_size, color, line_thickness, num_lines):
    """
    Create a single tile with randomized lines that can tile seamlessly.
    Mirrors any lines crossing the edges to ensure seamless tiling.
    """
    lines = []
    for _ in range(num_lines):
        # Start point
        x1, y1 = np.random.uniform(0, tile_size, 2)
        # End point
        x2, y2 = np.random.uniform(0, tile_size, 2)
        lines.append(((x1, y1), (x2, y2)))

        # If the line crosses a boundary, create a mirrored line on the opposite edge
        if x1 == x2 or y1 == y2:  # Simple case: horizontal or vertical lines
            continue  # No need to mirror perfectly horizontal or vertical lines
        # Mirror horizontally
        if x1 < tile_size / 2:
            lines.append(((x1 + tile_size, y1), (x2 + tile_size, y2)))
        else:
            lines.append(((x1 - tile_size, y1), (x2 - tile_size, y2)))
        # Mirror vertically
        if y1 < tile_size / 2:
            lines.append(((x1, y1 + tile_size), (x2, y2 + tile_size)))
        else:
            lines.append(((x1, y1 - tile_size), (x2, y2 - tile_size)))

    return lines

def draw_pattern(tile_size, num_tiles_x, num_tiles_y, color, line_thickness, num_lines):
    fig, ax = plt.subplots(figsize=(num_tiles_x, num_tiles_y))
    ax.set_aspect('equal')
    plt.axis('off')

    # Generate the base tile
    base_tile_lines = create_base_tile(tile_size, color, line_thickness, num_lines)

    # Repeat the base tile across the specified area
    for i in range(num_tiles_x):
        for j in range(num_tiles_y):
            for line in base_tile_lines:
                start, end = line
                plt.plot([start[0] + i * tile_size, end[0] + i * tile_size],
                         [start[1] + j * tile_size, end[1] + j * tile_size], color=color, linewidth=line_thickness)

    filename = generate_unique_filename_png()
    file_path = f"/content/{filename}"
    plt.savefig(file_path, bbox_inches='tight', pad_inches=0, dpi=100)
    plt.close(fig)

    return filename

# Parameters
tile_size = 100
num_tiles_x = 5
num_tiles_y = 5
color = "#6495ED"  # Example color
line_thickness = 2
num_lines = 10  # Number of lines in each tile

# Generate and save the pattern
pattern_filename = draw_pattern(tile_size, num_tiles_x, num_tiles_y, color, line_thickness, num_lines)

print(f"Saved pattern image as {pattern_filename}")
```

```python
def generate_parametric_sculpture_and_save_to_obj(filename="parametric_sculpture.obj"):
    # Define the sculpture geometry
    u = np.linspace(0, 2 * np.pi, 30)  # Reduced for simplicity
    v = np.linspace(0, np.pi, 30)  # Reduced for simplicity
    u, v = np.meshgrid(u, v)

    x = np.sin(v) * np.cos(u)
    y = np.sin(v) * np.sin(u)
    z = np.cos(v) + np.sin(u) * 3

    # Flatten the arrays and combine into a list of vertices
    vertices = np.vstack([x.flatten(), y.flatten(), z.flatten()]).T

    # Generate faces (triangles) based on the grid
    faces = []
    for i in range(len(u) - 1):
        for j in range(len(v) - 1):
            # Define vertices for two triangles (quad) per grid element
            v1 = i + j * len(u) + 1
            v2 = i + (j + 1) * len(u) + 1
            v3 = i + j * len(u) + 2
            v4 = i + (j + 1) * len(u) + 2
            faces.append([v1, v2, v3])
            faces.append([v2, v4, v3])

    # Write to OBJ file
    with open(f"/content/{filename}", "w") as file:
        for vertex in vertices:
            file.write(f"v {vertex[0]} {vertex[1]} {vertex[2]}\n")
        for face in faces:
            file.write(f"f {face[0]} {face[1]} {face[2]}\n")

    return filename

# Generate and save the sculpture to an OBJ file
filename = generate_parametric_sculpture_and_save_to_obj()
print(f"Sculpture saved to {filename}")
```
